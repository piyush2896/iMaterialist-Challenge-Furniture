### Some Data Exploration!

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from glob import glob
from random import shuffle
import os

%matplotlib inline
```

```python
def get_labels(root):
  unique_labels = os.listdir(root)
  return unique_labels

def get_train_files_and_labels(root):
  unique_labels = get_labels(root)
  img_files = []
  img_labels = []
  for i, label in enumerate(unique_labels):
    files = glob(os.path.join(root, label, '*'))
    img_files = img_files + files
    img_labels = img_labels + [i for j in range(len(files))]
  return img_files, img_labels, unique_labels

def get_shapes_lists(root, n_imgs=1.0):
  heights = []
  widths = []

  img_files, _, _ = get_train_files_and_labels(root)
  print('Taking {}% of total images for shapes'.format(int(n_imgs * 100)))
  n_imgs = int(len(img_files)* n_imgs)

  indices = np.random.randint(len(img_files), size=n_imgs)

  for index in indices:
    img = cv2.imread(img_files[index])
    heights.append(img.shape[0])
    widths.append(img.shape[1])
  return heights, widths

def plot_shapes_hist(root):
  heights, widths = get_shapes_lists(root, 0.1)

  plt.figure(0, figsize=(20, 10))
  plt.subplot(1, 2, 1)
  plt.hist(heights, bins=8)
  plt.xlabel('Counts')
  plt.ylabel('height')
  plt.title('Image Height vs N Images')
  plt.subplot(1, 2, 2)
  plt.hist(widths, bins=8)
  plt.xlabel('Counts')
  plt.ylabel('width')
  plt.title('Image Width vs N Images')
  plt.show()

def plot_some_imgs(root):
  img_files, labels, unique_labels = get_train_files_and_labels(root)

  print('Displaying 10 images from dataset')
  indices = np.random.randint(len(img_files), size=10)

  plt.figure(figsize=(20, 20))
  for i, index in enumerate(indices):
    plt.subplot(5, 2, i+1)
    plt.imshow(cv2.cvtColor(cv2.imread(img_files[index]), cv2.COLOR_BGR2RGB))
    plt.title('Label: ' + unique_labels[labels[index]])
    plt.axis('off')
  plt.show()
```

### Model Building Starts Here!

Transfer learning will play crucial roll in bumping up the performance of the model. Although the dataset is quite unique dataset. But there is a few ressemblance with imagenet. I'll be using the **InceptionResNet v2** as base model and will be adding a layer beyond it.
> A Dense output layer wiht 128 neurons and softmax activation

![InceptionResNet v2](https://1.bp.blogspot.com/-O7AznVGY9js/V8cV_wKKsMI/AAAAAAAABKQ/maO7n2w3dT4Pkcmk7wgGqiSX5FUW2sfZgCLcB/s1600/image00.png)

<a href = "https://ai.googleblog.com/2016/08/improving-inception-and-image.html">Improving Inception and Image Classification in TensorFlow</a> </div>



```python
from keras.applications.inception_resnet_v2 import InceptionResNetV2, preprocess_input
from keras.layers import Dense
from keras.models import Model
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input,
                                   shear_range=0.3, # Side Bend Images
                                   zoom_range=0.3,  # Zoom a little into images
                                   horizontal_flip=True) # Horizontally flip the images

valid_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = train_datagen.flow_from_directory(TRAIN_DATA_DIR,
                                                    shuffle=True,
                                                    batch_size=BATCH_SIZE,
                                                    target_size=IN_SIZE,
                                                    class_mode='categorical')

validation_generator = valid_datagen.flow_from_directory(VALID_DATA_DIR,
                                                         shuffle=False,
                                                         batch_size=BATCH_SIZE,
                                                         target_size=IN_SIZE,
                                                         class_mode='categorical')
```

```python
base_model = InceptionResNetV2(weights='imagenet',
                               include_top=False,
                               pooling='max') # Apply global average pooling 
```

```python
x = Dense(128, activation='softmax')(base_model.output)

model = Model(base_model.input, x)
```

```python
for layer in base_model.layer: layer.trainable = False


model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit_generator(train_generator,
                      train_generator.n // BATCH_SIZE,
                      epochs=1, 
                      workers=4, 
                      validation_data=validation_generator,
                      validation_steps=validation_generator.n // BATCH_SIZE)
```

```python
from keras.optimizers import Adam
import numpy as np
```

```python
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

test_generator = test_datagen.flow_from_directory(TEST_DATA_DIR,
                                                  shuffle=False,
                                                  batch_size=1,
                                                  target_size=IN_SIZE,
                                                  class_mode=None)
```

Making 1/3rd of the total layers to be trainable!!

```python
split = len(base_model.layer) - (len(base_model.layers) // 3)

for layer in base_model.layers[:split]: layer.trainable=False
for layer in base_model.layers[split:]: layer.trainable=True

opt = Adam(lr=1e-2)

model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
```

```python
model.fit_generator(train_generator,
                    train_generator.n // BATCH_SIZE,
                    epochs=3, 
                    workers=4, 
                    validation_data=validation_generator,
                    validation_steps=validation_generator.n // BATCH_SIZE)
```

```python
preds = model.predict_generator(test_generator, steps=test_generator.n)
np.save('preds.npy', preds)
```

```python
model_json = model.to_json()
with open('model.json', 'w') as json_file:
  json_file.write(model_json)

model.save_weights('model.h5')
```
